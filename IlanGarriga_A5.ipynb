```python
class song:
    def __init__(self, titol, autor, durada):
        self.titol = titol
        self.autor = autor
        self.durada = durada
    def info(self):
        segons = self.durada % 60
        print(f"--- Info de la Canco ---")
        print(f"Titol: {self.titol}")
        print(f"Autor: {self.autor}")
        print(f"Durada:{segons:02d} minuts ({self.durada} segons)")
class Playlist:
    def __init__(self, nom, song_list=None):
        self.nom = nom
        if song_list is None:
            self.song_list = []
        else:
            self.song_list = song_list
    def afegir_canco(self, song):
        self.song_list.append(song)
        print(f"S'ha afegit '{song.titol}' a la playlist '{self.nom}'.")
    def durada_total(self):
        total_segons = sum(s.durada for s in self.song_list)
        return f"{total_segons} segons"
    def canco_mes_llarga(self):
        if not self.song_list:
            return "La llista esta buida."
        mes_llarga = self.song_list[0]
        for canco in self.song_list:
            if canco.durada > mes_llarga.durada:
                mes_llarga = canco
        return f"La canco mes llarga es: {mes_llarga.titol}"
    def mostrar_playlist(self):
        print(f"\n PLAYLIST: {self.nom} ")
        for i, song in enumerate(self.song_list, 1):
            print(f"{i}  {song.titol}  {song.autor} ({song.durada}s)")
        print(f"Durada total: {self.durada_total()}")
        print("-" * 67)
canco1 = song("Never Gonna Give You Up", "Rick Astley", 273)
canco2 = song("Som Carles Kirk", "Catalan cover ", 273)
canco3 = song("Nose", "Ilan Garriga", 67)
lallista = Playlist("Preferides 2024")
lallista.afegir_canco(canco1)
lallista.afegir_canco(canco2)
lallista.afegir_canco(canco3)
lallista.mostrar_playlist()
print(lallista.canco_mes_llarga())
print(f"Temps total: {lallista.durada_total()}")
```

```python
class Personatge:
    def __init__(self, nom, vida, forca, nivell):
        self.nom = nom
        self.vida = vida
        self.forca = forca
        self.nivell = nivell
    def atacar(self, enemic):
        enemic.vida -= self.forca
        if enemic.vida < 0:
            enemic.vida = 0
        print(f"{self.nom} ataca a {enemic.nom}! Vida restant de {enemic.nom}: {enemic.vida}")
    def pujar_nivell(self):
        self.nivell += 1
        self.forca += 2
        self.vida += 5
        print(f"
 {self.nom} ha pujat al nivell {self.nivell}! (Força +2, Vida +5)")
    def esta_viu(self):
        return self.vida > 0
    def __str__(self):
        return f"Personatge: {self.nom} | Nivell: {self.nivell} | Vida: {self.vida} | Força: {self.forca}"
    def info(self):
        print(self.__str__())
p1 = Personatge("Tirador", 40, 10, 1)
p2 = Personatge("Kirk", 35, 8, 1)
torn = 1
while p1.esta_viu() and p2.esta_viu():
    print(f"\nTorn {torn}:")
    p1.atacar(p2)
    if p2.esta_viu():
        p2.atacar(p1)
    torn += 1
if p1.esta_viu():
    guanyador = p1
else:
    guanyador = p2
print(f"Guanyador: {guanyador.nom}")
guanyador.pujar_nivell()
print("\nEstadistiques finals del guanyador:")
guanyador.info()
```

```python
class SistemaEquacions:
    def __init__(self, matriu_A, vector_B):
        self.A = matriu_A
        self.B = vector_B         
        self.n = len(vector_B)
        self.solucio = []
    def dibuixar_matriu(self, m):
        for fila in m:
            valors = [round(v, 2) for v in fila[:-1]]
            resultat = round(fila[-1], 2)
            print(f"{valors} = {resultat}")
        print("-" * 15)
    def resoldre(self):
        n = self.n
        m = []
        for i in range(n):
            m.append(self.A[i] + [self.B[i]])
        print("Estat inicial:")
        self.dibuixar_matriu(m)
        for i in range(n):
            pivot = m[i][i]
            print(f"Pivot a la fila {i+1}: {pivot}")
            if abs(pivot) < 1e-10:
                print("Error sistema")
                return None
            for j in range(i, n+1):
                m[i][j] /= pivot
            for k in range(n):
                if k != i:
                    f = m[k][i]
                    for j in range(i, n+1):
                        m[k][j] -= f * m[i][j]
            print(f"Pas {i+1}:")
            self.dibuixar_matriu(m)
        self.solucio = [fila[n] for fila in m]
        return self.solucio
    def mostrar(self):
        if self.solucio:
            print("Solucions:")
            for i, v in enumerate(self.solucio):
                print(f"x{i+1} = {round(v, 2)}")
coefs = [
    [1, 1, 1],
    [0, 1, 0],
    [0, 0, 3]
]
res = [8, 4, 9]
sistema = SistemaEquacions(coefs, res)
sistema.resoldre()
sistema.mostrar()
```
